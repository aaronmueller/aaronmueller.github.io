```python
import torch
import time
```

```python
vocab_size = 10000
start_loop = time.process_time()
vocab_vector = []
for i in range(vocab_size):
    vocab_vector.append(0)
end_loop = time.process_time()

start_torch = time.process_time()
vocab_vector = torch.zeros(vocab_size)
end_torch = time.process_time()

print("Time taken w/ loop:", end_loop - start_loop)
print("Time taken w/ torch:", end_torch - start_torch)
```

```python
# initialize two vectors of random numbers
vec1 = torch.randn(vocab_size)
vec2 = torch.randn(vocab_size)
dot_prod = 0.0

# dot product via for loop
start_loop = time.process_time()
for i in range(vocab_size):
    dot_prod += vec1[i] * vec2[i]
end_loop = time.process_time()

# dot product via torch
start_torch = time.process_time()
vocab_vector = torch.dot(vec1, vec2)
end_torch = time.process_time()

print("Time taken w/ loop:", end_loop - start_loop)
print("Time taken w/ torch:", end_torch - start_torch)
```

```python
from tqdm import tqdm

num_iterations = 5000
num_epochs = 10

start_loop = time.process_time()
for i in range(num_epochs):
    for i in tqdm(range(num_iterations), leave=False):
        for i in range(vocab_size):
            dot_prod += vec1[i] * vec2[i]
end_loop = time.process_time()

print("Time taken w/ loop:", end_loop - start_loop)
```

```python
start_torch = time.process_time()
for i in range(num_epochs):
    for i in tqdm(range(num_iterations)):
        vocab_vector = torch.dot(vec1, vec2)
end_torch = time.process_time()

print("Time taken w/ torch:", end_torch - start_torch)
```
